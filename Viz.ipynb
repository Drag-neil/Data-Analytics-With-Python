```python
import seaborn as sns
import pandas as pd
```

```python
bd = pd.read_csv('cars.csv',encoding= 'ISO-8859-1')
```

```python
bd.head()
```

## Univariate

```python
#number = Histograms,boxplot
#Categorial = barplots  #Either variables having non numbers or a small range of numbers in which all the values are
```

```python
sns.displot(bd['hp'],bins=5);
```

```python
sns.boxplot(y='hp',data=bd)
```

```python
# Count frequency of categories
sns.countplot(x='gear',data=bd)
```

```python
# Shows density of data (Similar to kde plots) - Just another way
sns.violinplot(y='hp',data=bd);
```

## Bivariate Analysis

## Numeric - Numeric

```python
sns.jointplot(x="hp",y="disp",data=bd) #Num-Num - Scatterplot 
#Scatter plot , dot plots
```

```python
sns.lmplot(x='hp', y='disp', data=bd, fit_reg=True); #positive correlation since line goes from bottom to top 
```

## Categorical - Categorical

```python
sns.countplot(x="am", hue='vs',data=bd)#Cat-Cat - SidebySide barplot 
```

## Numeric - Categorical

```python
sns.boxplot(x='am',y='hp',data=bd) #Num-Cat - Boxplot
```
